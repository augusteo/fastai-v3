```python
from fastai import *
from fastai.vision import *
```

```python
classes = ['cheetah', 'geoffroys_cat', 'jaguar', 'leopard', 'ocelot']
```

```python
# only for kaggle kernel
input_path = Path('../input')
```

```python
for c in classes:
    print(c)
    folder = c
    file = c + '.csv'
    path = Path('data')
    dest = path/folder
    dest.mkdir(parents=True, exist_ok=True)
    download_images(input_path/file, dest, max_pics=400)
    verify_images(dest, delete=True, max_size=500)
```

```python
# view data
np.random.seed(42)
data = ImageDataBunch.from_folder(path, train=".", valid_pct=0.2,
     ds_tfms=get_transforms(), size=64, num_workers=0).normalize(imagenet_stats)
```

```python
data.classes
```

```python
data.show_batch(rows=3, figsize=(3,4))
```

```python
learn = cnn_learner(data, models.resnet50, metrics=error_rate)
```

```python
learn.fit_one_cycle(4)
```

```python
learn.save('stage-1')
```

```python
learn.unfreeze()
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(8, max_lr=slice(1e-4,1e-3))
```

```python
learn.save('stage-2')
```

# interpretation

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_confusion_matrix()
```

```python
np.random.seed(42)
clean_path = Path('data/cleaned.csv')
data = ImageDataBunch.from_csv(".", folder="data", valid_pct=0.2, csv_labels='data/cleaned.csv',
ds_tfms=get_transforms(), size=224, num_workers=4).normalize(imagenet_stats)
```

```python
learn = cnn_learner(data, models.resnet50, metrics=error_rate)
learn.fit_one_cycle(4)
```

```python
learn.unfreeze()
learn.lr_find()
learn.recorder.plot()
```

```python
learn.fit_one_cycle(40, max_lr=slice(1e-5,2e-4))
```

```python
learn.save('cleaned')
```

## analysis of cleaned model

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```
